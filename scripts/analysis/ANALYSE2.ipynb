```
# Import required libraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
df = pd.read_excel('indoor warehouse data.xlsx')

# Convert Date column to datetime if not already
df['Date'] = pd.to_datetime(df['Date'])

# Create future dates for prediction
last_date = df['Date'].max()
future_dates = pd.date_range(start='2025-01-01', end='2026-01-31', freq='MS')

# Fit a polynomial regression for Indoor data
X = (df['Date'] - df['Date'].min()).dt.days.values.reshape(-1, 1)
y_indoor_pkg = df['INDOOR Pkg'].values
y_indoor_cbm = df['INDOOR CBM'].values

# Create polynomial features
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit models for Indoor
model_indoor_pkg = LinearRegression().fit(X_poly, y_indoor_pkg)
model_indoor_cbm = LinearRegression().fit(X_poly, y_indoor_cbm)

# Create future X values
future_X = (future_dates - df['Date'].min()).days.values.reshape(-1, 1)
future_X_poly = poly.transform(future_X)

# Make predictions
indoor_pkg_pred = model_indoor_cbm.predict(future_X_poly)
indoor_cbm_pred = model_indoor_cbm.predict(future_X_poly)

# Create prediction dataframe for Indoor
indoor_pred_df = pd.DataFrame({
    'Date': future_dates,
    'Predicted_Packages': indoor_pkg_pred,
    'Predicted_CBM': indoor_cbm_pred
})

# Ensure no negative values
indoor_pred_df['Predicted_Packages'] = indoor_pred_df['Predicted_Packages'].clip(lower=0)
indoor_pred_df['Predicted_CBM'] = indoor_pred_df['Predicted_CBM'].clip(lower=0)

print("Indoor Warehouse Predictions (2025-01 to 2026-01):")
print(indoor_pred_df.to_string(index=False))
```

```
# Now let's do the same for Outdoor data
# Fit models for Outdoor
y_outdoor_pkg = df['OUTDOOR Pkg'].values
y_outdoor_cbm = df['OUTDOOR CBM'].values

# Fit models
model_outdoor_pkg = LinearRegression().fit(X_poly, y_outdoor_pkg)
model_outdoor_cbm = LinearRegression().fit(X_poly, y_outdoor_cbm)

# Make predictions
outdoor_pkg_pred = model_outdoor_pkg.predict(future_X_poly)
outdoor_cbm_pred = model_outdoor_cbm.predict(future_X_poly)

# Create prediction dataframe for Outdoor
outdoor_pred_df = pd.DataFrame({
    'Date': future_dates,
    'Predicted_Packages': outdoor_pkg_pred,
    'Predicted_CBM': outdoor_cbm_pred
})

# Ensure no negative values
outdoor_pred_df['Predicted_Packages'] = outdoor_pred_df['Predicted_Packages'].clip(lower=0)
outdoor_pred_df['Predicted_CBM'] = outdoor_pred_df['Predicted_CBM'].clip(lower=0)

print("Outdoor Warehouse Predictions (2025-01 to 2026-01):")
print(outdoor_pred_df.to_string(index=False))

# Create visualization
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
plt.title('Predicted Warehouse Volume (CBM) - 2025-2026')
plt.plot(indoor_pred_df['Date'], indoor_pred_df['Predicted_CBM'], label='Indoor CBM', marker='o')
plt.plot(outdoor_pred_df['Date'], outdoor_pred_df['Predicted_CBM'], label='Outdoor CBM', marker='o')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.title('Predicted Package Count - 2025-2026')
plt.plot(indoor_pred_df['Date'], indoor_pred_df['Predicted_Packages'], label='Indoor Packages', marker='o')
plt.plot(outdoor_pred_df['Date'], outdoor_pred_df['Predicted_Packages'], label='Outdoor Packages', marker='o')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# Import required libraries
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns

# Read and prepare data
df = pd.read_excel('indoor warehouse data.xlsx')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Analyze seasonality and trends for both Indoor and Outdoor
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Indoor Analysis
indoor_decompose_pkg = seasonal_decompose(df['INDOOR Pkg'], period=3, model='additive')
indoor_decompose_cbm = seasonal_decompose(df['INDOOR CBM'], period=3, model='additive')

# Plot Indoor Package decomposition
axes[0,0].plot(indoor_decompose_pkg.trend)
axes[0,0].set_title('Indoor Package Trend')
axes[0,0].tick_params(axis='x', rotation=45)

# Plot Indoor CBM decomposition
axes[0,1].plot(indoor_decompose_cbm.trend)
axes[0,1].set_title('Indoor CBM Trend')
axes[0,1].tick_params(axis='x', rotation=45)

# Outdoor Analysis
outdoor_decompose_pkg = seasonal_decompose(df['OUTDOOR Pkg'], period=3, model='additive')
outdoor_decompose_cbm = seasonal_decompose(df['OUTDOOR CBM'], period=3, model='additive')

# Plot Outdoor Package decomposition
axes[1,0].plot(outdoor_decompose_pkg.trend)
axes[1,0].set_title('Outdoor Package Trend')
axes[1,0].tick_params(axis='x', rotation=45)

# Plot Outdoor CBM decomposition
axes[1,1].plot(outdoor_decompose_cbm.trend)
axes[1,1].set_title('Outdoor CBM Trend')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Calculate correlation between Indoor and Outdoor usage
correlation_pkg = df['INDOOR Pkg'].corr(df['OUTDOOR Pkg'])
correlation_cbm = df['INDOOR CBM'].corr(df['OUTDOOR CBM'])

print("\nCorrelation Analysis:")
print("Correlation between Indoor and Outdoor Packages:", correlation_pkg)
print("Correlation between Indoor and Outdoor CBM:", correlation_cbm)

# Calculate monthly growth rates
df['Indoor_Pkg_Growth'] = df['INDOOR Pkg'].pct_change()
df['Indoor_CBM_Growth'] = df['INDOOR CBM'].pct_change()
df['Outdoor_Pkg_Growth'] = df['OUTDOOR Pkg'].pct_change()
df['Outdoor_CBM_Growth'] = df['OUTDOOR CBM'].pct_change()

print("\nMonthly Growth Rates (Average):")
print("Indoor Package Growth Rate:", df['Indoor_Pkg_Growth'].mean())
print("Indoor CBM Growth Rate:", df['Indoor_CBM_Growth'].mean())
print("Outdoor Package Growth Rate:", df['Outdoor_Pkg_Growth'].mean())
print("Outdoor CBM Growth Rate:", df['Outdoor_CBM_Growth'].mean())
```

```
# Perform advanced time series forecasting using SARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings('ignore')

# Prepare future dates
future_dates = pd.date_range(start='2025-01-01', end='2026-01-31', freq='MS')

# Function to fit SARIMA model and make predictions
def sarima_forecast(data, future_dates):
    model = SARIMAX(data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    results = model.fit()
    forecast = results.get_forecast(steps=len(future_dates))
    mean_forecast = forecast.predicted_mean
    return pd.Series(mean_forecast, index=future_dates)

# Make predictions for each metric
indoor_pkg_forecast = sarima_forecast(df['INDOOR Pkg'], future_dates)
indoor_cbm_forecast = sarima_forecast(df['INDOOR CBM'], future_dates)
outdoor_pkg_forecast = sarima_forecast(df['OUTDOOR Pkg'], future_dates)
outdoor_cbm_forecast = sarima_forecast(df['OUTDOOR CBM'], future_dates)

# Create forecast dataframes
indoor_forecast = pd.DataFrame({
    'Date': future_dates,
    'Predicted_Packages': indoor_pkg_forecast.values,
    'Predicted_CBM': indoor_cbm_forecast.values
})

outdoor_forecast = pd.DataFrame({
    'Date': future_dates,
    'Predicted_Packages': outdoor_pkg_forecast.values,
    'Predicted_CBM': outdoor_cbm_forecast.values
})

# Ensure no negative values and apply project completion adjustment
completion_factor = np.linspace(1, 0.2, len(future_dates))  # Gradual decrease to 20% by project end

indoor_forecast[['Predicted_Packages', 'Predicted_CBM']] = \
    indoor_forecast[['Predicted_Packages', 'Predicted_CBM']].clip(lower=0) * completion_factor.reshape(-1, 1)

outdoor_forecast[['Predicted_Packages', 'Predicted_CBM']] = \
    outdoor_forecast[['Predicted_Packages', 'Predicted_CBM']].clip(lower=0) * completion_factor.reshape(-1, 1)

print("Updated Indoor Warehouse Forecast (2025-01 to 2026-01):")
print(indoor_forecast.to_string(index=False))
print("\nUpdated Outdoor Warehouse Forecast (2025-01 to 2026-01):")
print(outdoor_forecast.to_string(index=False))

# Visualize updated forecasts
plt.figure(figsize=(15, 10))

# Plot CBM forecasts
plt.subplot(2, 1, 1)
plt.title('Updated Warehouse Volume Forecast (CBM) - 2025-2026')
plt.plot(indoor_forecast['Date'], indoor_forecast['Predicted_CBM'], 
         label='Indoor CBM', marker='o')
plt.plot(outdoor_forecast['Date'], outdoor_forecast['Predicted_CBM'], 
         label='Outdoor CBM', marker='o')
plt.legend()
plt.grid(True)

# Plot Package forecasts
plt.subplot(2, 1, 2)
plt.title('Updated Package Count Forecast - 2025-2026')
plt.plot(indoor_forecast['Date'], indoor_forecast['Predicted_Packages'], 
         label='Indoor Packages', marker='o')
plt.plot(outdoor_forecast['Date'], outdoor_forecast['Predicted_Packages'], 
         label='Outdoor Packages', marker='o')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```
# 1. Data Loading and Initial Cleaning
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_excel('WAREHOUSE INDOOR_OUTDOOR QTY.xlsx')
df['Date'] = pd.to_datetime(df['Date'])

# Handle missing or infinite values
df['INDOOR Pkg'] = df['INDOOR Pkg'].replace([np.inf, -np.inf], np.nan)
df['INDOOR CBM'] = df['INDOOR CBM'].replace([np.inf, -np.inf], np.nan)

# Remove rows with missing values
df_clean = df.dropna(subset=['INDOOR Pkg', 'INDOOR CBM'])

print("Data Cleaning Results:")
print(f"Original rows: {len(df)}")
print(f"Cleaned rows: {len(df_clean)}")
print("\nBasic Statistics for Cleaned Data:")
print("\nINDOOR Package Statistics:")
print(df_clean['INDOOR Pkg'].describe())
print("\nINDOOR CBM Statistics:")
print(df_clean['INDOOR CBM'].describe())
```

```
# 2. Multiple Validation Methods

# Method 1: Time Series Cross-Validation
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Prepare data for validation
y_pkg = df_clean['INDOOR Pkg'].values
y_cbm = df_clean['INDOOR CBM'].values

# Time series cross-validation
tscv = TimeSeriesSplit(n_splits=3)
hw_results_pkg = []
hw_results_cbm = []

for train_idx, test_idx in tscv.split(y_pkg):
    # Packages validation
    train_pkg = y_pkg[train_idx]
    test_pkg = y_pkg[test_idx]
    
    model_pkg = ExponentialSmoothing(train_pkg, seasonal_periods=3, trend='add', seasonal='add')
    fitted_pkg = model_pkg.fit()
    pred_pkg = fitted_pkg.forecast(len(test_pkg))
    
    hw_results_pkg.append({
        'mae': mean_absolute_error(test_pkg, pred_pkg),
        'rmse': np.sqrt(mean_squared_error(test_pkg, pred_pkg))
    })
    
    # CBM validation
    train_cbm = y_cbm[train_idx]
    test_cbm = y_cbm[test_idx]
    
    model_cbm = ExponentialSmoothing(train_cbm, seasonal_periods=3, trend='add', seasonal='add')
    fitted_cbm = model_cbm.fit()
    pred_cbm = fitted_cbm.forecast(len(test_cbm))
    
    hw_results_cbm.append({
        'mae': mean_absolute_error(test_cbm, pred_cbm),
        'rmse': np.sqrt(mean_squared_error(test_cbm, pred_cbm))
    })

# Calculate average metrics
avg_metrics_pkg = {
    'mae': np.mean([r['mae'] for r in hw_results_pkg]),
    'rmse': np.mean([r['rmse'] for r in hw_results_pkg])
}

avg_metrics_cbm = {
    'mae': np.mean([r['mae'] for r in hw_results_cbm]),
    'rmse': np.mean([r['rmse'] for r in hw_results_cbm])
}

print("Time Series Cross-Validation Results:")
print("\nPackages Metrics:")
print(f"Average MAE: {avg_metrics_pkg['mae']:.2f}")
print(f"Average RMSE: {avg_metrics_pkg['rmse']:.2f}")

print("\nCBM Metrics:")
print(f"Average MAE: {avg_metrics_cbm['mae']:.2f}")
print(f"Average RMSE: {avg_metrics_cbm['rmse']:.2f}")

# Calculate ratio stability
df_clean['CBM_per_Package'] = df_clean['INDOOR CBM'] / df_clean['INDOOR Pkg']
ratio_stats = df_clean['CBM_per_Package'].describe()

print("\nCBM per Package Ratio Statistics:")
print(ratio_stats)
```

```
# 3. Advanced Statistical Analysis and Cross-Validation

import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy import stats
import statsmodels.api as sm

# Prepare data for modeling
indoor_df = indoor_df.sort_values('Date')
X = np.arange(len(indoor_df)).reshape(-1, 1)
y_pkg = indoor_df['INDOOR Pkg'].values
y_cbm = indoor_df['INDOOR CBM'].values

# Time series cross-validation
tscv = TimeSeriesSplit(n_splits=3)
cv_results_pkg = []
cv_results_cbm = []

# Multiple model validation
def validate_models(X, y, cv):
    results = []
    for train_idx, test_idx in cv.split(X):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        # Linear Regression
        model = sm.OLS(y_train, sm.add_constant(X_train)).fit()
        y_pred = model.predict(sm.add_constant(X_test))
        
        # Calculate metrics
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        results.append({
            'MSE': mse,
            'MAE': mae,
            'R2': r2,
            'Test_Size': len(y_test)
        })
    return results

# Run validation for both Package and CBM
cv_results_pkg = validate_models(X, y_pkg, tscv)
cv_results_cbm = validate_models(X, y_cbm, tscv)

# Calculate average metrics
def print_cv_results(results, name):
    avg_mse = np.mean([r['MSE'] for r in results])
    avg_mae = np.mean([r['MAE'] for r in results])
    avg_r2 = np.mean([r['R2'] for r in results])
    
    print(f"\n{name} Cross-Validation Results:")
    print(f"Average MSE: {avg_mse:.2f}")
    print(f"Average MAE: {avg_mae:.2f}")
    print(f"Average R²: {avg_r2:.2f}")
    
print_cv_results(cv_results_pkg, "Package")
print_cv_results(cv_results_cbm, "CBM")

# Additional Validation Tests
# 1. Durbin-Watson test for autocorrelation
model_pkg = sm.OLS(y_pkg, sm.add_constant(X)).fit()
model_cbm = sm.OLS(y_cbm, sm.add_constant(X)).fit()

dw_pkg = sm.stats.stattools.durbin_watson(model_pkg.resid)
dw_cbm = sm.stats.stattools.durbin_watson(model_cbm.resid)

print("\nDurbin-Watson Test Results:")
print(f"Package DW statistic: {dw_pkg:.4f}")
print(f"CBM DW statistic: {dw_cbm:.4f}")

# 2. Breusch-Pagan test for heteroskedasticity
bp_pkg = sm.stats.diagnostic.het_breuschpagan(model_pkg.resid, model_pkg.model.exog)
bp_cbm = sm.stats.diagnostic.het_breuschpagan(model_cbm.resid, model_cbm.model.exog)

print("\nBreusch-Pagan Test Results:")
print(f"Package BP test p-value: {bp_pkg[1]:.4f}")
print(f"CBM BP test p-value: {bp_cbm[1]:.4f}")

# 3. Calculate prediction intervals
def calc_prediction_intervals(model, X, alpha=0.05):
    y_pred = model.predict(sm.add_constant(X))
    mse = np.sum(model.resid**2) / (len(X) - 2)
    
    X_with_const = sm.add_constant(X)
    var_pred = mse * (1 + np.diag(X_with_const @ np.linalg.inv(X_with_const.T @ X_with_const) @ X_with_const.T))
    
    t_value = stats.t.ppf(1 - alpha/2, len(X) - 2)
    pi = t_value * np.sqrt(var_pred)
    
    return y_pred - pi, y_pred + pi

pkg_lower, pkg_upper = calc_prediction_intervals(model_pkg, X)
cbm_lower, cbm_upper = calc_prediction_intervals(model_cbm, X)

print("\nPrediction Interval Ranges:")
print(f"Package PI width (avg): {np.mean(pkg_upper - pkg_lower):.2f}")
print(f"CBM PI width (avg): {np.mean(cbm_upper - cbm_lower):.2f}")

# Calculate forecast accuracy metrics
def calc_forecast_accuracy(y_true, y_pred):
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return {'MAPE': mape, 'RMSE': rmse}

pkg_accuracy = calc_forecast_accuracy(y_pkg, model_pkg.predict(sm.add_constant(X)))
cbm_accuracy = calc_forecast_accuracy(y_cbm, model_cbm.predict(sm.add_constant(X)))

print("\nForecast Accuracy Metrics:")
print(f"Package MAPE: {pkg_accuracy['MAPE']:.2f}%")
print(f"Package RMSE: {pkg_accuracy['RMSE']:.2f}")
print(f"CBM MAPE: {cbm_accuracy['MAPE']:.2f}%")
print(f"CBM RMSE: {cbm_accuracy['RMSE']:.2f}")
```

```
# 4. Seasonal Decomposition and Advanced Time Series Analysis
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, kpss
import matplotlib.pyplot as plt

# Set the date as index for time series analysis
ts_df = indoor_df.set_index('Date')

# Perform seasonal decomposition
def analyze_seasonality(series, name):
    # Decompose the series
    decomposition = seasonal_decompose(series, period=2, extrapolate_trend='freq')
    
    # Plot decomposition
    plt.figure(figsize=(15, 10))
    plt.subplot(411)
    plt.plot(series)
    plt.title(f'{name} - Original')
    plt.subplot(412)
    plt.plot(decomposition.trend)
    plt.title('Trend')
    plt.subplot(413)
    plt.plot(decomposition.seasonal)
    plt.title('Seasonal')
    plt.subplot(414)
    plt.plot(decomposition.resid)
    plt.title('Residual')
    plt.tight_layout()
    plt.show()
    
    return decomposition

# Analyze both series
pkg_decomp = analyze_seasonality(ts_df['INDOOR Pkg'], 'Packages')
cbm_decomp = analyze_seasonality(ts_df['INDOOR CBM'], 'CBM')

# Stationarity Tests
def check_stationarity(series, name):
    # ADF Test
    adf_result = adfuller(series.dropna())
    # KPSS Test
    kpss_result = kpss(series.dropna())
    
    print(f"\nStationarity Tests for {name}:")
    print("ADF Test:")
    print(f"p-value: {adf_result[1]:.4f}")
    print("\nKPSS Test:")
    print(f"p-value: {kpss_result[1]:.4f}")
    
# Run stationarity tests
check_stationarity(ts_df['INDOOR Pkg'], 'Packages')
check_stationarity(ts_df['INDOOR CBM'], 'CBM')

# Calculate volatility
def calculate_volatility(series, name):
    returns = series.pct_change().dropna()
    volatility = returns.std() * np.sqrt(12)  # Annualized volatility
    return volatility

pkg_vol = calculate_volatility(ts_df['INDOOR Pkg'], 'Packages')
cbm_vol = calculate_volatility(ts_df['INDOOR CBM'], 'CBM')

print("\nAnnualized Volatility:")
print(f"Packages: {pkg_vol:.2%}")
print(f"CBM: {cbm_vol:.2%}")

# Correlation Analysis
correlation = ts_df['INDOOR Pkg'].corr(ts_df['INDOOR CBM'])
print(f"\nCorrelation between Packages and CBM: {correlation:.4f}")

# Calculate rolling statistics
window = 3  # 3-month rolling window
rolling_mean_pkg = ts_df['INDOOR Pkg'].rolling(window=window).mean()
rolling_std_pkg = ts_df['INDOOR Pkg'].rolling(window=window).std()
rolling_mean_cbm = ts_df['INDOOR CBM'].rolling(window=window).mean()
rolling_std_cbm = ts_df['INDOOR CBM'].rolling(window=window).std()

# Plot rolling statistics
plt.figure(figsize=(15, 10))
plt.subplot(2,1,1)
plt.plot(ts_df.index, ts_df['INDOOR Pkg'], label='Original')
plt.plot(ts_df.index, rolling_mean_pkg, label=f'{window}-Month Rolling Mean')
plt.plot(ts_df.index, rolling_std_pkg, label=f'{window}-Month Rolling Std')
plt.title('Package Rolling Statistics')
plt.legend()

plt.subplot(2,1,2)
plt.plot(ts_df.index, ts_df['INDOOR CBM'], label='Original')
plt.plot(ts_df.index, rolling_mean_cbm, label=f'{window}-Month Rolling Mean')
plt.plot(ts_df.index, rolling_std_cbm, label=f'{window}-Month Rolling Std')
plt.title('CBM Rolling Statistics')
plt.legend()
plt.tight_layout()
plt.show()

# Calculate stability metrics
def calculate_stability_metrics(series):
    cv = series.std() / series.mean()  # Coefficient of variation
    range_ratio = (series.max() - series.min()) / series.mean()
    return {'CV': cv, 'Range_Ratio': range_ratio}

pkg_stability = calculate_stability_metrics(ts_df['INDOOR Pkg'])
cbm_stability = calculate_stability_metrics(ts_df['INDOOR CBM'])

print("\nStability Metrics:")
print("\nPackages:")
print(f"Coefficient of Variation: {pkg_stability['CV']:.4f}")
print(f"Range Ratio: {pkg_stability['Range_Ratio']:.4f}")
print("\nCBM:")
print(f"Coefficient of Variation: {cbm_stability['CV']:.4f}")
print(f"Range Ratio: {cbm_stability['Range_Ratio']:.4f}")
```

```
# Generate future predictions using the best performing model (Random Forest)
future_dates = pd.date_range(start='2025-01-01', end='2025-12-31', freq='M')
X_future = np.arange(len(indoor_df), len(indoor_df) + len(future_dates)).reshape(-1, 1)

# Generate predictions
future_pkg_pred_rf = rf_model_pkg.predict(X_future)
future_cbm_pred_rf = rf_model_cbm.predict(X_future)

# Create forecast DataFrame
forecast_rf_df = pd.DataFrame({
    'Date': future_dates,
    'Predicted_Packages': future_pkg_pred_rf,
    'Predicted_CBM': future_cbm_pred_rf
})

# Ensure no negative values
forecast_rf_df['Predicted_Packages'] = forecast_rf_df['Predicted_Packages'].clip(lower=0)
forecast_rf_df['Predicted_CBM'] = forecast_rf_df['Predicted_CBM'].clip(lower=0)

# Round the values
forecast_rf_df = forecast_rf_df.round(2)

print("\nMonthly Forecast using Random Forest Model:")
print(forecast_rf_df.to_string(index=False))

# Visualize the predictions
plt.figure(figsize=(15, 10))

# Plot Package predictions
plt.subplot(2, 1, 1)
plt.plot(indoor_df['Date'], indoor_df['INDOOR Pkg'], 'b-', label='Historical Packages')
plt.plot(forecast_rf_df['Date'], forecast_rf_df['Predicted_Packages'], 'r--', label='RF Predicted Packages')
plt.title('Package Forecast using Random Forest')
plt.legend()

# Plot CBM predictions
plt.subplot(2, 1, 2)
plt.plot(indoor_df['Date'], indoor_df['INDOOR CBM'], 'g-', label='Historical CBM')
plt.plot(forecast_rf_df['Date'], forecast_rf_df['Predicted_CBM'], 'r--', label='RF Predicted CBM')
plt.title('CBM Forecast using Random Forest')
plt.legend()

plt.tight_layout()
plt.show()

# Calculate confidence intervals using Random Forest's built-in feature
def predict_with_std(model, X):
    predictions = np.zeros((len(X), model.n_estimators))
    for i, estimator in enumerate(model.estimators_):
        predictions[:, i] = estimator.predict(X)
    mean = np.mean(predictions, axis=1)
    std = np.std(predictions, axis=1)
    return mean, std

# Get confidence intervals
_, pkg_std = predict_with_std(rf_model_pkg, X_future)
_, cbm_std = predict_with_std(rf_model_cbm, X_future)

# Add confidence intervals to the forecast
forecast_rf_df['Packages_CI_Lower'] = (forecast_rf_df['Predicted_Packages'] - 1.96 * pkg_std).clip(lower=0)
forecast_rf_df['Packages_CI_Upper'] = forecast_rf_df['Predicted_Packages'] + 1.96 * pkg_std
forecast_rf_df['CBM_CI_Lower'] = (forecast_rf_df['Predicted_CBM'] - 1.96 * cbm_std).clip(lower=0)
forecast_rf_df['CBM_CI_Upper'] = forecast_rf_df['Predicted_CBM'] + 1.96 * cbm_std

print("\nMonthly Forecast with Confidence Intervals:")
print(forecast_rf_df.round(2).to_string(index=False))
```
